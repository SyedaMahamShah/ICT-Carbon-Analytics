```python
import sys
import os
sys.path.append(os.path.abspath(".."))

from src.lifecycle_model import lifecycle_emissions
import matplotlib.pyplot as plt

result = lifecycle_emissions("server", "finland")

labels = ["Production", "Use Phase", "End of Life"]
values = [
    result["Production"],
    result["Use Phase"],
    result["End of Life"]
]

plt.bar(labels, values)
plt.title("Server Lifecycle Emissions (Finland)")
plt.ylabel("kg CO2e")
plt.show()
```

```python
countries = ["finland", "france"]

for country in countries:
    result = lifecycle_emissions("server", country)

    labels = ["Production", "Use Phase", "End of Life"]
    values = [
        result["Production"],
        result["Use Phase"],
        result["End of Life"]
    ]

    plt.figure()
    plt.bar(labels, values)
    plt.title(f"Server Lifecycle Emissions ({country.capitalize()})")
    plt.ylabel("kg CO2e")
    plt.xticks(rotation=20)
    plt.show()
```

```python
result = lifecycle_emissions("server", "finland")

production_share = result["Production"] / result["Total"] * 100
use_share = result["Use Phase"] / result["Total"] * 100
eol_share = result["End of Life"] / result["Total"] * 100

print(f"Production Share: {production_share:.1f}%")
print(f"Use Phase Share: {use_share:.1f}%")
print(f"End-of-Life Share: {eol_share:.1f}%")
print(f"Total: {(production_share + use_share + eol_share):.1f}%")
```

```python
labels = ["Production", "Use Phase", "End of Life"]
sizes = [
    production_share,
    use_share,
    eol_share
]

plt.figure()
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Lifecycle Emission Share (Server - Finland)")
plt.show()
```
